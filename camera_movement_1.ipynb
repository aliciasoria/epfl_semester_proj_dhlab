```python
import json
%matplotlib notebook 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = open('sfm_data_expanded.json')
sfm_data = json.load(data)
```

```python
sfm_data.keys()
```

```python
x_data = []
y_data = []
z_data = []
for p in range(len(sfm_data['extrinsics'])):
    x_data.append(sfm_data['extrinsics'][p]['value']['center'][0])
    y_data.append(sfm_data['extrinsics'][p]['value']['center'][1])
    z_data.append(sfm_data['extrinsics'][p]['value']['center'][2])
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
ax.plot3D(x_data, y_data, z_data, 'gray')

# Data for three-dimensional scattered points
ax.scatter3D(x_data, y_data, z_data)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Camera positions wrt. world coordinates')
plt.show()
```
